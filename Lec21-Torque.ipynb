# Torque

A mass $\vec m$ is moving with velocity $\vec v$ has momentum $\vec p$. There may be a force acting upon that mass $\vec F$. There is a random point Q, and the position vector between m and Q is $\vec r_Q$. Angular momentum is

$$\vec L_Q = \vec r_Q \times \vec p \quad \quad (1)$$

Its direction can be known by right hand rule

The torque

$$\vec \tau_Q =  \vec r_Q \times \vec F \quad \quad (2)$$

Similary, the direction can be known by right hand rule

$$\frac{d\vec L_Q}{dt} = \vec \tau_Q external \quad \quad (3)$$

The torque leads to the change of angular momentum. If there is no torque, there is no change in angular momentum. If you have a system of particles, if there is no net external torque, on the system as a whole, angular momentum will be conserved

Rotation about Q

$$|\tau_Q| = I_Q \alpha_Q \quad \quad (4)$$

$\alpha = \ddot \theta$ and $w_Q = \dot \theta$

$$|L_Q| = I_Q w_Q \quad \quad (5)$$

Rotation about stationary axis through center of mass has spin angular momentum

$$|L_{cm}| = I_{cm} w_{cm} \quad \quad (6)$$


### Example: Earth and Sun

An Earth m is going around the Sun at c. Position vector from c to m is $\vec r_c$, tangential velocity is $\vec v$, angular velocity w, and gravitational force is $\vec F$ points towards the center .

From equation 1

$$|L_c| = |\vec r_Q \times \vec p| = r_c mv = mR^2 w$$

There is no sine because the angle between r and v is 90 degree

We can get the same result with this equation 5

$$|L_Q| = I_Q w_Q = mR^2 w$$

Because the moment of inertia of mass m at distance R is $mR^2$

Equation 2 tells us that the torque is zero relative to point c because direction of r and F is always in the same direction when m spin around c. Angular momentum is only conserved (not changing) relative to c. If you choose any other points, the angular momentum is changing


### Example: Rotate Rod on Point not Center of Mass

A rotating rod with mass M, length l, center of mass c. It spins clockwise about point P with distance between point P and center of mass is d

$$|L_P| = I_Pw$$

A rod rotate at center of mass has moment of inertia $\frac{1}{12} Ml^2$ and from the parallel axis theorem,

$$|L_P| = (\frac{1}{12} Ml^2 + Md^2)w$$

It is rotating clockwise, so the direction is into the blackboard

Note that there will be force acting towards point P from the longer side. This is because one side of the rod has larger mass than another side. If the pivot is at center of mass, force from both end will cancel out, thus zero force

But since we take torque relative to point p, then position vector has length zero so no torque. 

$$|\vec\tau_p| = 0 \implies |\vec L_p| = 0$$

But if take torque relative at other point, there will be a force act towards point p, hence angular momentum not conserved



### Example: Rotate Rod on Center of Mass

Because rod rotate at center of mass (symmetry), there is no net force acted on center of mass, so torque relative to any point is zero

$$|\vec\tau_Q| = 0$$

Now we have the case that angular momentum of this rotating object regardless which point you choose. This called spin angular momentum. It is an intrinsic property of a spinning object

$$|L_{cm}| = I_cw = \frac{1}{12} Ml^2w$$


### Example: Impluse on a Rod

An impulse is applied horizontally on a point p from center of mass c. What will this object behave?

The intuition that the rod will rotate at any arbitrary point is wrong. The object must rotate about center of mass. 

Let say the rod does not rotate about center of mass. Then it means center of mass will be rotating. It is not allowed because the center of mass behaves like a single point in which all the mass is concentrated. After the horizontal hit, it is absurd that a single point will rotate

So, how nature digests the impulse is that the center of mass will move with velocity $v_{cm}$ and the point hit by the impulse will rotate with $w_{cm}$

The center of mass acts like a point source, so

$$\vec F = m\vec a_{cm}$$

$$Impulse = F \Delta t = m\vec a_{cm} \Delta t = mv_{cm}$$

$$\implies v_{cm} = \frac{Impulse}{m}$$

Velocity of cm only depends on the impulse and mass, and doesn't depend on where you hit the rod

Next question, what is the angular momentum and torque relative to point c. Torque is r cross F, so certainly torque is non-zero. If torque is non zero, then angular momentum is changing

$$\vec \tau \neq 0$$

$$L_{before} = 0$$

$$L_{after} = I_{p} w_{cm}$$

Though velocity of cm is insensitive of where you hit, the angular velocity is very sensitive

If we take point p as our origin, then

$$\vec \tau_p = 0$$

$L_p$ is conserved, 0 before, 0 after. Solution of taking point p or cm as origin give same answer for angular velocity
