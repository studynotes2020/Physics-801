# Angular Momentum

Momentum, $\vec p = m\vec v$

Angular momentum

$$\vec L_Q = \vec r_Q \times \vec p = ( \vec r_Q \times \vec v)m$$

$r_Q$ is the position vector between Q and the object. I can choose any point Q arbitrarily

Magnitude of angular momentum

$$|L_Q| = mv r_Q \sin\theta$$

$r_Q \sin\theta$ is r from Q perpendicular to p, $r_Q \bot $

In this case, direction of r is vertical, direction of p is horizontal, then the direction of angular momentum is into the blackboard

If position vector r is heading the same direction as p, $\sin\theta = 0$, the angular momentum is zero

Angular momentum is not intrinsic property for a moving object. If a mass moving with a velocity, then it has momentum. But, angular momentum depends on the origin that you choose. In the example, we can choose an origin that gives angular momentum goes into the blackboard


### Example

Suppose I throw an object with projection motion. At time zero, it is at C, and at time t, its position vector is $\vec r_c$. Starting velocity is v, and there is a force $mg$ acting downwards

At t = 0, position vector has no length, angular momentum $\vec L_c = 0$

In this case, velocity is changing, hence angular momentum is changing

Here an example where velocity is changing but angular momentum is not changing

The Earth, m, going around the Sun at c. The position vector $\vec r_c$ is from c to m. The Earth has tangential velocity $\vec v$. Speed doesn't change but velocity does change

$$|\vec L_c| = m |\vec r_c \times \vec v| = mrv$$

The angle between v and r is 90, throughout the rotation. So the angular momentum wrt to point c is not changing.

Suppose we pick another point Q which is on the track of rotation. Then the angular momentum wrt Q is changing. When m pass through Q, angular momemtum is zero, and at the other points it is non-zero

Angular momentum is conserved only about the special case of point c

$$\vec L_Q = \vec r_Q \times \vec p$$

$$\frac{d\vec L_Q}{dt} = \frac{d\vec r_Q}{dt} \times \vec p + \vec r_Q \times \frac{d\vec p}{dt}$$

$\frac{d\vec r_Q}{dt} = \vec v$  is always in the same direction as momentum, so that cross term is zero

$\frac{d\vec p}{dt} = \vec F$ is force

$$\frac{d\vec L_Q}{dt} = \vec r_Q \times \vec F = \vec \tau_Q$$

which is called torque. This equation tells us if there is torque, the angular momentum is changing

The gravitational pull of m from c, $\vec F$ is in the same direction with $r_c$, so there is no torque relative to point c

However if we take Q, or some other points, the torque is not zero, hence the angular momentum is changing


### Example

A disk with center c, mass M, radius r, is rotating with angular velocity w. We want to know its angular momentum

Take a point mass $m_i$ at position $r_{i}$ wrt c, tangential velocity $\vec v_i$, then $r_i \times p$, the angular momentum is pointing out of blackboard

What is the magnitude of the disk as a whole?

Relative to c

$$L_i = m_i r_i v_i$$

The velocity might be different and depends on its position, but they all have the same angular velocity, so we want to change velocity to angular velocity with $v = wr$ 

$$L_i = m_i r_i^2 w$$

This is for the little mass element. For the entire disk

$$L_{disk} = w\sum_i m_i r_i^2 = I_c w$$

$I = \sum_i m_i r_i^2$

The remarkable thing is that if you have a rotation about the center of mass, then angular momentum relative to other points is the same. This is only true when the rotation is about the center of mass. It is called the spin angular momentum

The spin angular momentum is an intrinsic property of an object regardless of which point you choose. When an object is spinning about its center of mass, we no longer need to specify the point chosen as origin. We can now talk about the angular momentum

If we want to talk about the orbital angular momentum of the Earth, we better take it relative to the point c. Otherwise, it would be changing in time. It is only uniquely defined if you take that special point c where the angular momentum is not changing



### Experiment

A man is holding two mass, each with 1.8 kg. The turntable together with the man is about 75 kg. The man will be rotating clockwise about his center of mass, so we can talk about the intrinsic angular momentum of this rotating system. 

While rotating, the man pulls his arm in and push out, it does not cause a net torque on the system. So the angular momentum for a spinning object must be conserved. However, when he pulls the arms in, his moment of inertia goes down. If $I$ goes down and $L$ remain constant, then the angular velocity must go up

$$L = Iw$$

Take the rotating object as cylinder with radius 20cm. When the man pulls his arm all in $I = \frac12 MR^2 = 1.5$

When the man stretches his arm with distance 90cm, the moment of inertia changes. From parallel axis theorem, $I' = I + md^2 = 1.5 + 2 * 1.8 * (0.9)^2 = 4.5$

So in prediction, if moment of inertia goes down by a factor of 3, then angular velocity must go up by a factor of 3



## Angular Momentum of a System

Take two particles in a system, the force between them cancel out. So if take any point Q, the angular momentum wrt to point Q is zero

Angular Momentum of a System of many particles will not change if there is no net external torque on that system, because all the internal torques cancel out, because action equals minus reaction

Angular Momentum of a System is conserved if no external torque


### Astrophysics

When a star shrinks, its radius goes down, its moment of inertia goes down, therefore its angular velocity goes up

Once the nuclear fuel of a star is consumed, there are three end products of the dead star that is left over

1. White dwarf with radius 10^4 km, mass = 0.5 mass of Sun. Density is 10^6 g/cm^3

2. Neutron star radius about 10 km, M is 1.5 mass of Sun, density is 10^14 g/cm^3

3. Blackhole, radius is 0, mass larger than 3 mass of Sun, density is infinitely high

Whether it ends up in white dwarf, neutron star or blackhole depends on the mass of the star that collapsed, when the nuclear fuel was gone

To collapse into neutron star, the star will need to start off at least 10x solar masses, so our Sun will not collapse into neutron star but white dwarf

Why is there no possibility between 0.5 mass (white dwarf) and 1.5 mass (neutron star)? It is explained in quantum mechanics. And for blackhole, we need to know general relativity

When a star collapses, two things happen. First, there is a huge amount of gravitational potential energy released in the form of kinetic energy and ultimately convert to heat and radiation
