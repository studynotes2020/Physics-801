# Rotating Objects

A rotating object with a velocity v, angular velocity w, distance to the origin is R, angle has increased by theta

$$v = wR = \dot\theta R$$

Take first derivative

$$a_{tangent} = \dot w R = \ddot\theta R = \alpha R$$

$\alpha$ is the angular acceleration in $rad/s^2$

Do not confuse tangential acceleration with centripetal acceleration

To convert linear motion to angular motion

$x \to \theta$

$v \to w$

$a \to \alpha$

$x = x_0 + v_0 + \frac12 at^2 \to \theta = \theta_0 + w_0 + \frac12 \alpha t^2$

$v = v_0 + at \to w = w_0 + \alpha t$


### Example

There a disk with origin C, radius R, mass m, rotating with w. How much kinetic energy of rotation is stored in the disk

Take a little mass element $m_i$ with radius $r_i$,

$$K_i = \frac12 m_i v_i^2 = \frac12 m_i w^2 r_i^2$$

Nice thing about writing in angular velocity is that for all points of the disk, it is the same whereas the velocity is not. Velocity closer to the center is low, further from the center is high. But omega is all the same

$$K_{disk} = \frac{w^2}{2}\sum m_i r_i^2$$

Call this $\sum m_i r_i^2 = I$ moment of inertia

$$K_{disk} = \frac12 I_c w^2$$

The subscript c means rotate at axis c

$m \to I$

There is a table for moment of inertia

For disk $I_c = \frac12 mR^2$

Solid sphere: $I = \frac25 mR^2$

Rod: $I = \frac{1}{12}ml^2$

Moment of inertia for rotating at different axis is different


### Rotating Disk

Center C, rotating at axis l, another parallel axis l'. Distance between l and l' is d

Parallel Axis Theorem

$$I_{l'} = I_l + md^2$$

Notice that $I_{l'}$ always larger than $I_l$


Perpendicular Axis Theory

A thin plate (work for this theory). For x, y, z axis

$$I_z = I_x + I_y$$


### Example: Special Car

A car goes down hill. $m = 10^3$ with speed 4 m/s. The bottom of valley is p with h = 500m. When reach point p,

$$K_p = \frac12 mv^2 = \frac12 \times 10 ^3 \times 4^2 = 8000 J$$

At the top of the hill

$$mgh = 5 \times 10^6 J$$

All those potential energy will be lost by heat if applying brake. 

Can we salvage some energy? Yes, install a flywheel, and convert some gravitational potential energy to the rotational energy of the flywheel

Flywheel disk with R = 0.5m, M = 200kg, $I = \frac12 MR^2 = 25 kg\ m^2$

We want to convert all the potential energy to it

$\frac12 Iw^2 = 5 \times 10^6 \implies w = 632 rad/s$ and $f = \frac{w}{2\pi} \approx 100Hz$

The slow down is not because of the brake dissipitating heat but convertion of gravitational potential energy to rotational energy of the flywheel


### Example: Flywheels

R = 2.4m, m = 85e3 kg, f = 6 Hz. With this, we can calculate I, and w, then $K_{rot} = 2 \times 10^8J$

They could convert this rotational energy to magnetic energy. They could dump this magnetic energy back to the flywheels


### Crab Pulsar

Crab pulsar is radiating out P = 6e31 Watt, which comes from its rotational energy

Rotation period of neutron star in the Crab Nebula, T = 0.0335028583 sec. Tomorrow, it will be slowing down by 36.4 nsec. The loss of energy is 6e3 J/sec which is the power recorded in terms of x-rays, gamma rays and other forms of energy. All the energy it radiates is at the expense of rotational kinetic energy
